## Useful Functions - Preprocessing

```python
from datetime import datetime # to strip datetime & use the inbuilt classes & objects

def print_summary(pandas_data_frame):
    """
    Function to print the summary of a Pandas data frame, 
    1 - Shape
    2 - Head
    3 - Information : data types of columns, number of entries, space used up
    4 - Describe : mean, 25%ile, 50%ile, 75%ile, minimum, maximum for each NUMERICAL COLUMN
    """
    print("======================= Shape =======================")
    print(pandas_data_frame.shape)
    print("\n======================= Head =======================")
    print(pandas_data_frame.head())
    print("\n======================= Info =======================")
    print(pandas_data_frame.info())
    print("\n======================= Describe =======================")
    print(pandas_data_frame.describe(include='all'))
    return

def print_missing_values_info(pandas_data_frame):
    """
    Function to print information of missing values in a Pandas data frame. NAs
    """
    missing_values = pd.DataFrame(pandas_data_frame.isna().sum())
    missing_values = missing_values.reset_index()
    missing_values.columns = ['variables', 'n_missing_values']
    missing_values['missing_percentage'] = missing_values['n_missing_values']*100/pandas_data_frame.shape[0]
    print("======================= Shape =======================")
    print(pandas_data_frame.shape)
    print("\n======================= Missing Values Info =======================")
    print(missing_values)
    return

def is_number(s):
    """
    Function to check if a given value can be converted to a valid float number
    """
    try:
        float(s)
        return True
    except ValueError:
        return False

def is_datetime(s):
    """
    Function to check if the given string can be stripped into the specified datetime format or not
    Takes care of anomalies in 'pickup_datetime' column
    """
    try:
        datetime_stripped = datetime.strptime(s, "%Y-%m-%d %H:%M:%S %Z")
        return True
    except:
        return False

def column_to_numeric_type(pandas_data_frame, colname):
    """
    Function to convert a column to numeric type by removing the non-numeric cell observations
    colname - string type
    """
    initial_n_rows = pandas_data_frame.shape[0]
    pandas_data_frame = pandas_data_frame[pandas_data_frame[colname].apply(lambda x: is_number(x))]
#     if pandas_data_frame[col].dtype is np.dtype('float64'):
    pandas_data_frame[colname] = pandas_data_frame[colname].apply(pd.to_numeric)
        # can also use the following
        # pandas_data_frame[col] = pandas_data_frame[col].astype('float')
    after_n_rows = pandas_data_frame.shape[0]
    n_rows_removed = initial_n_rows - after_n_rows
    print("Number of Rows removed = " + str(n_rows_removed) + "\n")
    return pandas_data_frame
#     return np.nan, False

def strip_pickup_datetime(pandas_data_frame, datetime_colname='pickup_datetime'):
    """
    Function to remove non-datetime types from the datetime columns & strip the datetime into datetime object type
    Then the year, month, day, weekday, hour, min, sec can be accessed using
    df['pickup_datetime'].loc[0].year / month / day / hour / minute / second / weekday()
    datetime_colname - string type
    """
    initial_n_rows = pandas_data_frame.shape[0]
    pandas_data_frame = pandas_data_frame[pandas_data_frame[datetime_colname].apply(lambda x: is_datetime(x))]
    pandas_data_frame[datetime_colname] = pandas_data_frame[datetime_colname].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S %Z"))
    after_n_rows = pandas_data_frame.shape[0]
    n_rows_removed = initial_n_rows - after_n_rows
    print("Number of Rows removed = " + str(n_rows_removed) + "\n")
    return pandas_data_frame

def drop_rows_with_0s_NAs(pandas_data_frame, columns_list):
    """
    Function to convert all 0s to np.nan & remove all the resulting NAs and already present NAs
    columns_list - list type
    """
    pandas_data_frame = pandas_data_frame.replace(0, np.nan)
    initial_n_rows = pandas_data_frame.shape[0]
    for col in columns_list:
        pandas_data_frame.dropna(subset=[col], inplace=True, axis='index')
    after_n_rows = pandas_data_frame.shape[0]
    n_rows_removed = initial_n_rows - after_n_rows
    print("Number of Rows removed = " + str(n_rows_removed) + "\n")
    print_missing_values_info(pandas_data_frame)
    return pandas_data_frame

def preprocess_data(pandas_data_frame):
    """
    Function to perform all the above preprocessing functions on a pandas_data_frame
    """
    pandas_data_frame = column_to_numeric_type(pandas_data_frame, 'fare_amount')
    pandas_data_frame = strip_pickup_datetime(pandas_data_frame)
    pandas_data_frame = drop_rows_with_0s_NAs(pandas_data_frame, pandas_data_frame.columns)
    # Changing the datatype of passenger_count to integer from float
    pandas_data_frame['passenger_count'] = pandas_data_frame['passenger_count'].astype('int')
    # Resetting the index numbers
    pandas_data_frame = pandas_data_frame.reset_index(drop=True)
    return pandas_data_frame
```

```python
def box_plot(y_axis_cols_list, x_axis_data_list, subplot_nrows, subplot_ncols, plot_heading='Box Plot'):
    """
    Plots boxplots for the columns in y_axis_cols_list & respective data in x_axis_data_list.
    Things to take care of:
    len(y_axis_cols_list) = len(x_axis_data_list) = subplot_ncols * subplot_nrows
    that means, if subplot_ncols = 6, subplot_nrows = 1, then the 
    number of columns for y_axis_cols_list & data in x_axis_data_list should be equal to 1*6 = 6
    y_axis_cols_list - string list
    x_axis_data_list - pandas dataframe list
    subplot_ncols, subplot_nrows - integer type
    """
    #Dividing the plotting area into subplots & storing the positions in ax matrix.
    #Each subplot can be accessed as indices of the matrix
    #Setting image size
    #plt.figure(figsize=(14,6))
    # squeeze = False parameter is being used because we are using only 1 row for subplots, so to avoid single axes indexing, squeeze = False
    fig, ax = plt.subplots(nrows = subplot_nrows, ncols = subplot_ncols, squeeze=False, figsize = (20,8))
    #Heading to the plot
    plot_title = plt.suptitle(plot_heading, fontsize = 'x-large')
    #Fits the plot into the specified figure size
    plt.tight_layout()
    #Plotting boxplots in the subplot area for each column
    for i in range(subplot_nrows):
        for j in range(subplot_ncols):
            list_index = subplot_ncols * i + j
            sns.boxplot(y=y_axis_cols_list[list_index], data=x_axis_data_list[list_index], ax=ax[i][j])
    fig.subplots_adjust(top = 0.9, right=0.9)
    plt.show()
    return

def hist_plot(y_axis_cols_list, x_axis_data_list, bins_list, subplot_nrows, subplot_ncols, plot_heading='Histogram Plot'):
    """
    Plots histogram plots for the columns in y_axis_cols_list & respective data in x_axis_data_list.
    Things to take care of:
    len(y_axis_cols_list) = len(x_axis_data_list) = len(bins_list) = subplot_ncols * subplot_nrows
    that means, if subplot_ncols = 6, subplot_nrows = 1, then the 
    number of columns for y_axis_cols_list & data in x_axis_data_list & integers in bins_list should be equal to 1*6 = 6
    y_axis_cols_list - string list
    x_axis_data_list - pandas dataframe list
    bins_list - integer list
    subplot_ncols, subplot_nrows - integer type
    """
    #Dividing the plotting area into subplots & storing the positions in ax matrix.
    #Each subplot can be accessed as indices of the matrix
    #Setting image size
    #plt.figure(figsize=(14,6))
    # squeeze = False parameter is being used because we are using only 1 row for subplots, so to avoid single axes indexing, squeeze = False
    fig, ax = plt.subplots(nrows = subplot_nrows, ncols = subplot_ncols, squeeze=False, figsize = (20,6))
    #Heading to the plot
    plot_title = plt.suptitle(plot_heading, fontsize = 'x-large')
    #Fits the plot into the specified figure size
    plt.tight_layout()
    #Plotting histogram plots in the subplot area for each column
    for i in range(subplot_nrows):
        for j in range(subplot_ncols):
            list_index = subplot_ncols * i + j
            data = x_axis_data_list[list_index]
            col = y_axis_cols_list[list_index]
            bins = bins_list[list_index]
            sns.distplot(data[col], ax=ax[i][j], bins=bins)
    fig.subplots_adjust(top = 0.9, right=0.9)
    plt.show()
    return

def bar_plot(categorical_list, subplot_nrows, subplot_ncols, plot_heading = 'Bar Plot', \
             pandas_data_frame = clean_training_data, plotting_method = 'mean'):
    """
    
    """
    #Dividing the plotting area into subplots & storing the positions in ax matrix.
    #Each subplot can be accessed as indices of the matrix
    #Setting image size
    #plt.figure(figsize=(14,6))
    # squeeze = False parameter is being used because we are using only 1 row for subplots,
    # so to avoid single axes indexing, squeeze = False
    fig, ax = plt.subplots(nrows = subplot_nrows, ncols = subplot_ncols, squeeze=False, figsize = (20,6))
    #Heading to the plot
    plot_heading = plot_heading + " : " + plotting_method
    plot_title = plt.suptitle(plot_heading, fontsize = 'x-large')
    #Fits the plot into the specified figure size
    plt.tight_layout()
    #Plotting barplots in the subplot area for each column
    for i in range(subplot_nrows):
        for j in range(subplot_ncols):
            list_index = subplot_ncols * i + j
            x_cat = categorical_list[list_index]
            group_temp = pandas_data_frame.groupby(by=x_cat)
#             temp1 = group_temp.count()
#             temp1['x_labels'] = temp1.index
#             temp1 = temp1[['onroad_distance', 'x_labels']]
#             x_axis = np.array(temp1['x_labels'])
            if plotting_method == 'mean':
                y_axis = np.array(group_temp.fare_amount.mean())
                x_axis = np.array(group_temp.fare_amount.mean().index)
            elif plotting_method == 'sum':
                y_axis = np.array(group_temp.fare_amount.sum())
                x_axis = np.array(group_temp.fare_amount.sum().index)
            elif plotting_method == 'count':
                y_axis = np.array(group_temp.fare_amount.count())
                x_axis = np.array(group_temp.fare_amount.count().index)
            sns.barplot(x=x_axis, y=y_axis, ax=ax[i][j])
    fig.subplots_adjust(top = 0.9, right=0.9)
    plt.show()
    return
```

## Useful Functions - Removing Outliers

```python
def find_outlier_range(pandas_data_frame_column):
    """
    Function to find outliers using IQR of a pandas column
    Lower range = Q1 - 1.5*IQR
    Upper range = Q3 + 1.5*IQR
    """
    q1 = np.quantile(np.array(pandas_data_frame_column), 0.25)
    q3 = np.quantile(np.array(pandas_data_frame_column), 0.75)
    iqr = q3 - q1
    lower_range = q1 - 1.5*iqr
    upper_range = q3 + 1.5*iqr
    outlier_range = upper_range - lower_range
    print("Lower Range of Outliers : " + str(lower_range))
    print("Upper Range of Outliers : " + str(upper_range))
    print("Outlier Range : " + str(outlier_range))
    return lower_range, upper_range

numerical_data_range = {
    'fare_amount' : [1, 100],
    'pickup_longitude' : [-74.8, -72.8],
    'pickup_latitude' : [39.45, 41.45],
    'dropoff_longitude' : [-74.8, -72.8],
    'dropoff_latitude' : [39.45, 41.45],
    'passenger_count' : [1, 6],
    'amnt_per_km' : [1, 20],
    'amnt_per_hr' : [20, 250],
    'onroad_distance' : [50, 100000000000000],
    'onroad_time' : [60, 1000000],
    'aerial_distance' : [10, 100000000000000]
}
numerical_data_range = pd.DataFrame(numerical_data_range)

def is_in_range(num, col_name):
    """
    Function to check if a given value is within the prescribed range of that column or not
    num - float or integer type
    col_name - string type
    """
    lower_range = numerical_data_range[col_name].loc[0]
    upper_range = numerical_data_range[col_name].loc[1]
    if num >= lower_range and num <= upper_range:
        return True
    return False

def remove_outliers(cols_list, pandas_data_frame):
    """
    Function to remove the rows containing outlier values using manually stored range
    col_list - list of strings of columns names
    """
    initial_n_rows = pandas_data_frame.shape[0]
    for col in cols_list:
        pandas_data_frame = pandas_data_frame[pandas_data_frame[col].apply(lambda x: is_in_range(x, col))]
    after_n_rows = pandas_data_frame.shape[0]
    n_rows_removed = initial_n_rows - after_n_rows
    print("Number of Rows removed = " + str(n_rows_removed) + "\n")
    return pandas_data_frame

def is_in_iqr_range(num, col_name, pandas_data_frame):
    """
    Function to check if a given value is within the IQR range of that column or not
    num - float or integer type
    col_name - string type
    """
    q1 = np.quantile(np.array(pandas_data_frame[col_name]), 0.25)
    q3 = np.quantile(np.array(pandas_data_frame[col_name]), 0.75)
    iqr = q3 - q1
    lower_range = q1 - 1.5*iqr
    upper_range = q3 + 1.5*iqr
    if num >= lower_range and num <= upper_range:
        return True
    return False

def remove_outliers_iqr(cols_list, pandas_data_frame):
    """
    Function to remove the rows containing outlier values using IQR Range
    col_list - list of strings of columns names
    """
    initial_n_rows = pandas_data_frame.shape[0]
    for col in cols_list:
        pandas_data_frame = pandas_data_frame[pandas_data_frame[col].apply(lambda x: is_in_iqr_range(x, col, pandas_data_frame))]
    after_n_rows = pandas_data_frame.shape[0]
    n_rows_removed = initial_n_rows - after_n_rows
    print("Number of Rows removed = " + str(n_rows_removed) + "\n")
    return pandas_data_frame

def preprocess_data_2(pandas_data_frame):
    """
    Performs all of the above functions
    """
    numerical_data_cols_list = ['fare_amount', 'pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude', 'passenger_count']
    pandas_data_frame = remove_outliers(pandas_data_frame=pandas_data_frame, cols_list=numerical_data_cols_list)
    pandas_data_frame = pandas_data_frame.reset_index(drop=True)
    return pandas_data_frame
```

```python
def find_onroad_distance(pandas_data_frame, coords_col_name='coords'):
    """
    Finds the ON-ROAD DISTANCE between the origin coordinates & destination coordinates
    1. Uses Google Maps API
    2. A Google Cloud Platform API Key is needed
    3. Install googlemaps library - !pip install -U googlemaps
    4. Result of the API is in JSON Format. Extracting data from it using conventional Python tools
    5. Taking distance value in meter
    6. Taking duration value in seconds
    """
    for i in range(len(pandas_data_frame[coords_col_name])):
        coord = pandas_data_frame[coords_col_name][i]
        origin = coord[0]
        dest = coord[1]
        #time = time.mktime(year=2019,month=8,day=)
        distance_results = gmaps.distance_matrix(origin, dest, mode='driving')
        on_road_distance = distance_results['rows'][0]['elements'][0]['distance']['value']
        on_road_time = distance_results['rows'][0]['elements'][0]['duration']['value']
        pandas_data_frame['onroad_distance'].loc[i] = on_road_distance
        pandas_data_frame['onroad_time'].loc[i] = on_road_time
    return pandas_data_frame

def find_onroad_distance_2(coord):
    """
    Finds the ON-ROAD DISTANCE between the origin coordinates & destination coordinates
    1. Uses Google Maps API
    2. A Google Cloud Platform API Key is needed
    3. Install googlemaps library - !pip install -U googlemaps
    4. Result of the API is in JSON Format. Extracting data from it using conventional Python tools
    5. Taking distance value in meter
    6. Taking duration value in seconds
    7. Doesn't use loops, instead works with apply() function and returns a tuple, so FASTER
    """
    origin = coord[0]
    dest = coord[1]
    #time = time.mktime(year=2019,month=8,day=)
    distance_results = gmaps.distance_matrix(origin, dest, mode='driving')
    return distance_results
#     on_road_distance = distance_results['rows'][0]['elements'][0]['distance']['value']
#     on_road_time = distance_results['rows'][0]['elements'][0]['duration']['value']
#     return on_road_distance, on_road_time

def find_aerial_dist(coord):
    """
    Function to find the aerial distance between the origin & destination coordinates
    Vincenty Formula is used for higher accuracy
    1. geopy library is used - !pip install geopy
    2. Distance is stored in meter
    """
    origin = coord[0]
    dest = coord[1]
    aerial_dist = geopy.distance.distance(origin, dest).m
    return aerial_dist

def fill_datetime_features(pandas_data_frame, datetime_col_name='pickup_datetime', weekday_col_name='weekday', month_col_name='winter_month', year_col_name='year', hour_col_name='rush_hour'):
    """
    Function to extract features from the datetime timestamp
    1. Year
    2. Month
    3. Weekday
    4. Hour
    Stored in new columns
    """
    for i in range(len(pandas_data_frame[datetime_col_name])):
        timestamp = pandas_data_frame[datetime_col_name].loc[i]
        wday = timestamp.weekday()
        pandas_data_frame['wday'].loc[i] = wday
        if (wday <= 4): # It IS A WEEKDAY : Monday - 0, Tuesday - 1,.... Friday - 4
            pandas_data_frame[weekday_col_name].loc[i] = 1
        else: # It is NOT A WEEKDAY : Saturday - 5, Sunday - 6
            pandas_data_frame[weekday_col_name].loc[i] = 0
        hr = timestamp.hour
        pandas_data_frame['hr'].loc[i] = hr
        if (hr in range(12,17) or hr in range(21,24) or hr in range(0, 2)):
            # Time adjusted from NY time to UTC for RUSH HOUR
            pandas_data_frame[hour_col_name].loc[i] = 1
        else:
            pandas_data_frame[hour_col_name].loc[i] = 0
        mnth = timestamp.month # 0-Jan, 1-Feb, 11-Dec : Winter seasons
        pandas_data_frame['mnth'].loc[i] = mnth
        if (mnth in range(2,11)): # NOT WINTER season
            pandas_data_frame[month_col_name].loc[i] = 0
        else: # IS WINTER season
            pandas_data_frame[month_col_name].loc[i] = 1
        pandas_data_frame[year_col_name].loc[i] = timestamp.year
    return pandas_data_frame

def set_winter_month(mnth):
    if (mnth in range(3, 12)):
        return 0 # NOT WINTER MONTH
    else:
        return 1 # WINTER MONTH

def set_weekday(wday):
    if (wday in range(0,5)):
        return 1 # IS WEEKDAY
    else:
        return 0 # NOT WEEKDAY

def set_late_night(hr): # NYC time is UTC -4hr, so latenight hours are 12AM - 3AM. UTC - 0400 to 0659
    if (hr in range(4, 7)):
        return 1
    else:
        return 0

def fill_new_features(pandas_data_frame):
    """
    Function to perform all of the above functions on a dataframe
    """
    pandas_data_frame.loc[:,'aerial_distance'] = pandas_data_frame.loc[:,'coords'].apply(lambda coord: find_aerial_dist(coord))
    pandas_data_frame.loc[:,'wday'] = pandas_data_frame.loc[:,'pickup_datetime'].apply(lambda dtime: dtime.weekday())
    pandas_data_frame.loc[:,'mnth'] = pandas_data_frame.loc[:,'pickup_datetime'].apply(lambda dtime: dtime.month)
    pandas_data_frame.loc[:,'hr'] = pandas_data_frame.loc[:,'pickup_datetime'].apply(lambda dtime: dtime.hour)
    pandas_data_frame.loc[:,'year'] = pandas_data_frame.loc[:,'pickup_datetime'].apply(lambda dtime: dtime.year)
    pandas_data_frame.loc[:,'winter_month'] = pandas_data_frame.loc[:,'mnth'].apply(lambda mnth: set_winter_month(mnth))
    pandas_data_frame.loc[:,'weekday'] = pandas_data_frame.loc[:,'wday'].apply(lambda wday: set_weekday(wday))
    pandas_data_frame.loc[:,'nght_hr'] = pandas_data_frame.loc[:,'hr'].apply(lambda hr: set_late_night(hr))
    pandas_data_frame.drop(columns=['pickup_datetime', 'pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude'], inplace=True, axis='column')
    return pandas_data_frame

def fill_gmaps_features(pandas_data_frame):
    temp_df = pd.DataFrame(pandas_data_frame.loc[:,'coords'].apply(lambda coord: find_onroad_distance_2(coord)))
    pandas_data_frame.loc[:,'onroad_distance'] = temp_df.loc[:,'coords'].apply(lambda coord: coord[0])
    pandas_data_frame.loc[:,'onroad_time'] = temp_df.loc[:,'coords'].apply(lambda coord: coord[1])
    return pandas_data_frame
```

### 4.2.2 Dependency Analysis between Categorical Variables - Chi-Square Test
scipy library

*p-value threshold* assumption - **0.05**

**Our test is significant (i.e. we reject the null hypothesis) if we get a p-value below our threshold.**

Null Hypothesis: Ho - The pair of categorical variables are **independent** i.e. there is **NO RELATIONSHIP**

Alternate Hypothesis: H1 - The pair of categorical variables are **not independent**

https://towardsdatascience.com/running-chi-square-tests-in-python-with-die-roll-data-b9903817c51b

```python
from scipy import stats
```

```python
clean_training_data
```

```python
paired_features = [(i, j) for i in categorical_columns for j in categorical_columns]
```

```python
def chi_square_test(pandas_data_frame=clean_training_data, categorical_features_list=categorical_columns):
    calc_p_vals = []
    for feature in paired_features:
        if feature[0] != feature[1]:
            data = pd.crosstab(clean_training_data[feature[0]], clean_training_data[feature[1]])
            chi2_stat, p_val, deg_of_freedom, result = stats.chi2_contingency(data)
            calc_p_vals.append(p_val.round(2))
        else:
            calc_p_vals.append(-1)
    print("=========== P-Values Calculated ===========")
    print(calc_p_vals)
    return calc_p_vals
```

Coefficient of Determination - R-squared - R2
The coefficient of determination is a key output of regression analysis. It is interpreted as the proportion of the **variance in the dependent variable** that is predictable from the **independent variables**.

The coefficient of determination is the 
**square of the correlation (r) between predicted y scores and actual y scores**;
thus, it ranges from **0 to 1**.

With linear regression, the coefficient of determination is also equal to the square of the correlation between x and y scores.

1. An R2 of 0 means that the dependent variable cannot be predicted from the independent variable.
2. An R2 of 1 means the dependent variable can be predicted without error from the independent variable.
3. An R2 between 0 and 1 indicates the extent to which the dependent variable is predictable. An R2 of 0.10 means that 10 percent of the variance in Y is predictable from X; an R2 of 0.20 means that 20 percent is predictable; and so on.

#### Removing IQR Outliers

```python
def is_in_iqr_range(num, col_name, pandas_data_frame):
    """
    Function to check if a given value is within the IQR range of that column or not
    num - float or integer type
    col_name - string type
    """
    q1 = np.quantile(np.array(pandas_data_frame[col_name]), 0.25)
    q3 = np.quantile(np.array(pandas_data_frame[col_name]), 0.75)
    iqr = q3 - q1
    lower_range = q1 - 1.5*iqr
    upper_range = q3 + 1.5*iqr
    if num >= lower_range and num <= upper_range:
        return True
    return False

def remove_outliers_iqr(cols_list, pandas_data_frame):
    """
    Function to remove the rows containing outlier values using IQR Range
    col_list - list of strings of columns names
    """
    initial_n_rows = pandas_data_frame.shape[0]
    for col in cols_list:
        pandas_data_frame = pandas_data_frame[pandas_data_frame[col].apply(lambda x: is_in_iqr_range(x, col, pandas_data_frame))]
    after_n_rows = pandas_data_frame.shape[0]
    n_rows_removed = initial_n_rows - after_n_rows
    print("Number of Rows removed = " + str(n_rows_removed) + "\n")
    return pandas_data_frame
```
